```python
import numpy as np
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn; seaborn.set()  # set plot style
```

```python
!head -4 data/diamonds.csv
```

```python
diamonds = pd.read_csv('data/diamonds.csv')
prices = np.array(diamonds['price'])
carats = np.array(diamonds['carat'])
```

### Q1 (2 points): Print the mean, standard deviation, minimum, maximum, 25th percentile, median, and 75th percentile of the prices array.

```python
print("="*40)
print("Mean price:       ", prices.mean())
print("Standard deviation:", prices.std())
print("Minimum price:    ", prices.min())
print("Maximum price:    ", prices.max())
print("="*40)
print("25th percentile:   ", np.percentile(prices, 25))
print("Median:            ", np.median(prices))
print("75th percentile:   ", np.percentile(prices, 75))
print("="*40)
```

### Q2 (2 points): Plot the histogram of the prices array.

```python
plt.hist(prices, 10)
plt.title('Price Distribution')
plt.xlabel('price')
plt.ylabel('number');
```

### Q3 (2 points): Reproduce the following output.

```python
print("Number of diamonds with price bigger than 1800:  ", np.sum(prices > 1800))
priceLarge=(prices>1800)
print("Average carat for diamonds with price bigger than 1800:  ", np.mean(carats[priceLarge]))
```

### Q4 (4 points): Reproduce the following scatter plot. The red circles represent a random subsample of the data.

```python
indices = np.random.choice(len(prices), 20, replace=False)
```

```python
select_price= prices[indices]
```

```python
select_carat= carats[indices]
```

```python
logPrice=np.log2(prices)
logCarat=np.log2(carats)

plt.scatter(logPrice, logCarat,alpha=0.3)
plt.xlabel('log2(prices)')
plt.ylabel('log2(carats)')
plt.scatter(np.log2(select_price), np.log2(select_carat),
            facecolor='none', s=200, edgecolors='r');
```
